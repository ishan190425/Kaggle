```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
LE = LabelEncoder()
enc = OneHotEncoder()
```

```python
data = pd.read_csv("/kaggle/input/titanic/train.csv")
data["Sex"] = LE.fit_transform(data.Sex)
data['Pclass'] = data['Pclass'].astype('float').values
data['Age'] = data['Age'].astype('float').values
data['SibSp'] = data['SibSp'].astype('float').values
data['Parch'] = data['Parch'].astype('float').values
data = data.drop(columns = "Cabin")
data = data.drop(columns = "Embarked")
data = data.drop(columns = "Ticket")
data = data.drop(columns = "Name")
data = data.drop(columns = "PassengerId")
data = data.drop(columns ='Fare')
data = data.fillna(-1)
y = data.Survived
x = data.drop(columns = "Survived")
```

```python
data
```

```python
X_train, X_test, y_train,y_test = train_test_split(x,y,test_size = .2, random_state = 0)
```

```python
X_train = sc.fit_transform(X_train) #applies feature scaling
X_test = sc.fit_transform(X_test)
X_train
```

```python
ann = tf.keras.models.Sequential()
```

```python
#add the input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation ='relu')) #input is the units (Number of nuerons in the first Layer) 
#(input layer is number of parameters) + hidden layers
#activation functions
#relu - rectifier function
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation ='relu')) #second layer, can change nuerons 
```

```python
ann.add(tf.keras.layers.Dense(units = 1, activation ='sigmoid')) #output layer, plus prob of 1, softmax when classifi
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy']) #using stoachstic Gradient Descent, #loss must be binary 
#when its a binary func
#categorical_crossentropy
```

```python
ann.fit(X_train,y_train,batch_size = 32, epochs = 1000)
```

```python
dataTest = pd.read_csv("/kaggle/input/titanic/test.csv")
dataTest = dataTest.drop(columns = "Cabin")
dataTest = dataTest.drop(columns = "Ticket")
dataTest = dataTest.drop(columns = "Name")
dataTest["Pclass"] = LE.fit_transform(dataTest.Pclass)
dataTest["Sex"] = LE.fit_transform(dataTest.Sex)
dataTest['Embarked'] = LE.fit_transform(dataTest.Embarked)
dataTest
```

```python
dataT = pd.DataFrame()
dataT["PassengerId"] = dataTest.PassengerId
dataTest['Pclass'] = dataTest['Pclass'].astype('float').values
dataTest['Age'] = dataTest['Age'].astype('float').values
dataTest['SibSp'] = dataTest['SibSp'].astype('float').values
dataTest['Parch'] = dataTest['Parch'].astype('float').values
dataTest = dataTest.drop(columns = 'PassengerId')
dataTest = dataTest.drop(columns ='Fare')
dataTest = dataTest.drop(columns = "Embarked")
dataTest = sc.fit_transform(dataTest)
dataT['Survived'] = ann.predict(dataTest).round(0)
dataT = dataT.fillna(0)
dataT['Survived'] = dataT['Survived'].astype('int')
```
